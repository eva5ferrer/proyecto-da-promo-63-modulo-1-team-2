```python
import random
```

```python
def pregunta_aleatoria(lista_preguntas, usadas):
    disponibles = [p for p in lista_preguntas if p["pregunta"] not in usadas]
    if not disponibles:
        return None
    return random.choice(disponibles)

#Sirve para: elegir una pregunta al azar y evitar que se repitan las peguntas en la partida
```

```python
def mostrar_pregunta(data):
    #Muestra el texto de la pregunta y las opciones A, B y C.
    print(data["pregunta"])
    for letra in ("A", "B", "C"):
        print(f"  {letra}) {data['opciones'][letra]}")
```

```python
def pedir_respuesta():
    #Pide al usuario A, B o C y valida la entrada.
    #Devuelve la letra en mayÃºscula.
    
    while True:
        r = input("Elige A, B o C: ").strip().upper()
        if r in ("A", "B", "C"):
            return r
        print(" Entrada invÃ¡lida. Escribe A, B o C.")
```

```python
def jugar_una_partida(quiz_geo):
    preguntas = quiz_geo["preguntas"]
    usadas = set()

    correctas = 0
    incorrectas = 0

    print(" QUIZ DE GEOGRAFÃA (A/B/C)\n")
    print("Reglas: ganas con 5 correctas o pierdes con 3 fallos.\n")

    while correctas < 5 and incorrectas < 3:
        data = pregunta_aleatoria(preguntas, usadas)

        if data is None:
            usadas.clear()
            data= preguntas_aleatoria(preguntas, usadas)

        usadas.add(data["pregunta"])

        mostrar_pregunta(data)
        r = pedir_respuesta()

        if r == data["correcta"]:
            correctas += 1
            print(" Correcto\n")
        else:
            incorrectas += 1
            letra = data["correcta"]
            texto = data["opciones"][letra]
            print(f" Incorrecto. Era {letra}) {texto}\n")

        print(f" Correctas: {correctas} |  Incorrectas: {incorrectas}\n")

    if correctas == 5:
        print(" Â¡HAS GANADO! Llegaste a 5 respuestas correctas.")
    else:
        print(" HAS PERDIDO. Llegaste a 3 respuestas incorrectas.")
```

```python
def preguntar_repetir():
   
    # Pregunta al usuario si quiere volver a jugar.
    #Acepta 'si' devuelve true o 'no' devuelve false.
    
    while True:
        r = input("Â¿Quieres volver a jugar? (si / no): ").strip().lower()

        if r == "si":
            return True
        if r == "no":
            return False

        print("âŒ Responde solo con 'si' o 'no'.")
```

```python
def iniciar_juego(quiz_geo):
    print("ğŸ² BIENVENIDO AL QUIZ DE GEOGRAFÃA")

    while True:
        jugar_una_partida(quiz_geo)
        if not preguntar_repetir():
            print("\nGracias por jugar. Â¡Hasta la prÃ³xima!")
            break
```
