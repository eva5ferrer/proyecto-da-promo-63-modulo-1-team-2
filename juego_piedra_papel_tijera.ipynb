## Juego Piedra, Papel y Tijera

```python
import random #importamos librer√≠a random que la usaremos m√°s adelante
opciones = ["piedra", "papel", "tijera"] #listamos las opciones disponibles para jugar, que usaremos durante todo el c√≥digo
```

```python
# Aqu√≠ creamos la funci√≥n para que el jugador 1 (usuario) pueda elejir entre pidra, papel, tijera o parar de jugar:
def jugada_usuario():
    while True:
        eleccion_jugador = input("""Para poder jugar tienes que elegir entre: piedra, papel o tijera. 
Si te aburrimos o ya est√°s embarcando y quieres parar de jugar, escribe "salir".""").strip().lower()
        if eleccion_jugador == "salir": #para evitar errores de espacios extra y may√∫sculas
            print("Entendido...no quieres jugar m√°s. ¬°Hasta pronto y buen viaje! ‚úàÔ∏è ")
            return "salir" #que nos devuelva "salir", para usarlo m√°s adelante
        elif eleccion_jugador not in opciones:
            print("""Ups, te has equivocado, esta opci√≥n no es v√°lida ‚ùå. ¬°Prueba de nuevo!
Recuerda que las opciones son: piedra, papel o tijera. Si quieres parar de jugar, escribe "salir".""")
            continue #aqu√≠ no devuelve nada
        print(f"üë§ Has escogido: {eleccion_jugador}.")
        return eleccion_jugador #que nos devuelva la eleccion_jugador, para usarlo m√°s adelante
```

```python
# Aqu√≠ creamos la funci√≥n para que el ordenador seleccione aleatoriamente una de las tres opciones (piedra, papel o tijera)
def jugada_ordenador():
    eleccion_ordenador = random.choice(opciones)
    print(f"üíª La jugada del ordenador es: {eleccion_ordenador}.")
    return eleccion_ordenador
```

```python
def determinar_ganador(jugada_usuario, jugada_ordenador):
    if jugada_usuario == jugada_ordenador:
        return "empate"
    elif (jugada_usuario =="piedra" and jugada_ordenador=="tijera") or\
         (jugada_usuario=="papel" and jugada_ordenador=="piedra") or\
         (jugada_usuario=="tijeras" and jugada_ordenador=="papel"):
        return "usuario"
    else:
        return "ordenador"
```

```python
def preguntar_repetir():
   
    # Pregunta al usuario si quiere volver a jugar.
    #Acepta 'si' devuelve true o 'no' devuelve false.
    
    while True:
        r = input("¬øQuieres volver a jugar? (si / no): ").strip().lower()

        if r == "si":
            return True
        if r == "no":
            return False

        print(" Por favor, responde solo con 'si' o 'no'.")
```

```python
def iniciar_partida():
    print("""¬°Hola!üëã ¬øEst√°s aburrido y quieres entretenerte un rato jugando al piedra, papel o tijera? 
Est√°s en el sitio id√≥neo. Sigue las instrucciones que te mostramos a continuaci√≥n.""")
    puntos_usuario= 0 # se crean los contadores a cero (tipo int)
    puntos_ordenador= 0


    while puntos_usuario <3 and puntos_ordenador <3: # aqui se definen los puntos necesarios para ganar, si ambos tienen menos de 3, cuando llegue a 3 se detiene el bucle.
        jugada = jugada_usuario() #le preguntamos al usuario qu√© quiere jugar

        if jugada == "salir": #paramos el bucle si el usuario escribe salir 
            break
        jugada_pc = jugada_ordenador() #turno para el ordenador
        resultado = determinar_ganador(jugada, jugada_pc) #como los dos han jugado, determinamos el ganador de la ronda

        if resultado =="usuario":
            puntos_usuario += 1
            print("Sigue as√≠. ¬°Has ganado la ronda! ü•≥")
        elif resultado == "ordenador":
            puntos_ordenador +=1
            print ("Oh... te ha ganado el ordenador esta vez üò¢")
        else:
            print("Esto est√° re√±ido, esta ronda ha sido un empate ‚öñÔ∏è")
        print(f"Puntuacion del usuarioüë§: {puntos_usuario} | Puntuaci√≥n del ordenador üíª: {puntos_ordenador}") #Esto sale siempre que se juegue una ronda
        print("-"*30)

    if puntos_usuario == 3: #cuando el usuario o el ordenador llega a 3 puntos, no entra en el bucle y determinamos ganador
        print ("Enhorabuena. ¬°Has ganado el juego! üéâ")
    elif puntos_ordenador == 3:
        print("Otra vez ser√°...el ordenador ha ganado el juego ü´Ç ")


```
