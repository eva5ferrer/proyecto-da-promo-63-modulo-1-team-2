## Juego Piedra, Papel y Tijera

```python
import random #importamos librería random que la usaremos más adelante
opciones = ["piedra", "papel", "tijera"] #listamos las opciones disponibles para jugar, que usaremos durante todo el código
```

```python
# Aquí creamos la función para que el jugador 1 (usuario) pueda elejir entre pidra, papel, tijera o parar de jugar:
def jugada_usuario():
    while True:
        eleccion_jugador = input("Para poder jugar tienes que elegir: piedra, papel o tijera. Si quieres parar de jugar, escribe 'salir'.").strip().lower()
        if eleccion_jugador == "salir": #para evitar errores de espacios extra y mayúsculas
            print("Entendido, no quieres jugar más. Hasta pronto!")
            return "salir" #que nos devuelva "salir", para usarlo más adelante
        elif eleccion_jugador not in opciones:
            print("""Ups, te has equivocado, esta opción no es válida, prueba de nuevo!
Recuerda que las opciones son: piedra, papel o tijera. Si quieres parar de jugar, escribe "salir".""")
            continue #aquí no devuelve nada
        print(f"Has escogido: {eleccion_jugador}.")
        return eleccion_jugador #que nos devuelva la eleccion_jugador, para usarlo más adelante
```

```python
jugada_usuario()
```

```python
# Aquí creamos la función para que el ordenador seleccione aleatoriamente una de las tres opciones (piedra, papel o tijera)
def jugada_ordenador():
    eleccion_ordenador = random.choice(opciones)
    print(f"La jugada del ordenador es: {eleccion_ordenador}.")
    return eleccion_ordenador
```

```python
jugada_ordenador()
```

```python
def determinar_ganador(jugada_usuario, jugada_ordenador):
    if jugada_usuario == jugada_ordenador:
        return "empate"
    elif (jugada_usuario =="piedra" and jugada_ordenador=="tijera") or\
         (jugada_usuario=="papel" and jugada_ordenador=="piedra") or\
         (jugada_usuario=="tijeras" and jugada_ordenador=="papel"):
        return "usuario"
    else:
        return "ordenador"
```

```python
def iniciar_juego():
    puntos_usuario= 0 # se crean los contadores a cero (tipo int)
    puntos_ordenador= 0


    while puntos_usuario <3 and puntos_ordenador <3: # aqui se definen los puntos necesarios para ganar, si ambos tienen menos de 3, cuando llegue a 3 se detiene el bucle.
        jugada = jugada_usuario() #le preguntamos al usuario qué quiere jugar

        if jugada == "salir": #paramos el bucle si el usuario escribe salir 
            break
        jugada_pc = jugada_ordenador() #turno para el ordenador
        resultado = determinar_ganador(jugada, jugada_pc) #como los dos han jugado, determinamos el ganador de la ronda

        if resultado =="usuario":
            puntos_usuario += 1
            print("Sigue así, has ganado la ronda!")
        elif resultado == "ordenador":
            puntos_ordenador +=1
            print ("Oh... te ha ganado el ordenador esta vez!")
        else:
            print("Esto está reñido, esta ronda ha sido un empate.")
        print(f"puntuacion al usuario: {puntos_usuario} / ordenador: {puntos_ordenador}") #Esto sale siempre que se juegue una ronda
        print("-"*30)

    if puntos_usuario == 3: #cuando el usuario o el ordenador llega a 3 puntos, no entra en el bucle y determinamos ganador
        print ("Enhorabuena, has ganado el juego!")
    elif puntos_ordenador == 3:
        print("Otra vez será, el ordenador ha ganado el juego :(")
```

```python
iniciar_juego() #aqui he escrito piedra y salir
```
